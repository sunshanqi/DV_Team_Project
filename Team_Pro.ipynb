```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
# Load data
adults_df = pd.read_csv('adult.data', header = None)
adults_df.columns = ["age", "workclass", "fnlwgt", "education", "education_num", "marital_status", "occupation",
                       "relationship","race","sex","capital_gain","capital_loss","hours_per_week","native_country","salary"]
adults_df.head()
```

```python
# Overview the data types
adults_df.info()
```

```python
print('Data frame shape: ', adults_df.shape)
```

```python
# Summary statistics
adults_df.describe()
```

```python
# Overview of target
sns.countplot(adults_df['salary'])
```

```python
# Display percentage of missing value of each feature
adults_df.isnull().sum()/len(adults_df)
```

```python
# Display scatter and distribution plots
sns.pairplot(adults_df)
plt.title('PairPlot')
plt.show()
```

```python
# Display the correlation matrix
corr_matrix = adults_df.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, vmax=1.0, vmin=-1.0,annot = True, annot_kws={"size": 15}, )
plt.title('Correlation Among Features')
plt.show()
```

```python
# Display box plot
plt.figure(figsize=(20, 10))
adults_df.boxplot()
plt.title('Box Ploting')
plt.show()
```
